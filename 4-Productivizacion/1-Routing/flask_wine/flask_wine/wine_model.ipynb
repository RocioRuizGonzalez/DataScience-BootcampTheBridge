```python
import numpy as np
import pandas as pd
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.metrics import accuracy_score
```

```python
wine = load_wine()
```

```python
data = pd.DataFrame(data= np.c_[wine['data'], wine['target']],
                     columns= wine['feature_names'] + ['target'])
data.head()
```

```python
import sqlite3
con = sqlite3.connect("wines.db")
data.to_sql("TRAINING", con)
```

```python
con.execute("SELECT * FROM TRAINING").fetchall()
```

```python
"CREATE TABLE IF NOT EXISTS PREDICTIONS (column_name datatype, column_name datatype);"

con.execute()
```

```python
data.target.unique()
```

```python
X_train = data[:-20]
X_test = data[-20:]

y_train = X_train.target
y_test = X_test.target

X_train = X_train.drop('target',1)
X_test = X_test.drop('target',1)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
print("accuracy_score: %.2f"
      % accuracy_score(y_test, y_pred))
```

```python
X = pd.concat([X_train, X_test])
y = pd.concat([y_train, y_test])

from sklearn.model_selection import cross_val_score
acc_cv = cross_val_score(clf, X, y, cv=5)
print("Performance in CV 5 folds: {:.2%}+-{:.2%}".format(np.mean(acc_cv), np.std(acc_cv)))
```

```python
np.random.seed(0)
X.sample(1).to_json(orient='index')
```
