## 1. Importa las librerias necesarias

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler, PolynomialFeatures

from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.svm import SVC, SVR
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline

from sklearn.metrics import mean_absolute_error, accuracy_score
```

## 2. Carga los datos

```python
df = pd.read_csv('data/Euro_2012_stats_TEAM.csv')
df
```

## 3. Obtener la matriz de correlación con TODAS las variables. ¿Hay algunas columnas que estén muy correlacionadas?

```python
df.dtypes
```

```python
ls_obj = df.dtypes[df.dtypes == object].index
pct_cols = df[ls_obj].applymap(lambda x: '%' in x).any()
ls_pct_cols = pct_cols[pct_cols].index
ls_pct_cols
```

```python
df_prep = df.copy()
df_prep.drop(columns=['Team'], inplace=True)
df_prep.loc[:, ls_pct_cols] = df_prep.loc[:, ls_pct_cols].apply(lambda x: x.str.replace('%', '')).astype(float)
df_prep
```

```python
df_prep.dtypes
```

```python
plt.figure(figsize=(5, 10))
sns.heatmap(df_prep.corr()[['Goals']], vmin=-1, vmax=1, annot=True, annot_kws={"size":13})
```

## 4. ¿Existe algún tipo de relación entre la precisión de disparo a puerta (Shooting Accuracy) con respecto a los goles? Muestra en una gráfica (pintando la línea en la distribución de ambas variables) y usando Regresión Lineal, Regresión no Lineal con polinomio y SVR la relación entre ambos

```python
sns.regplot(data=df_prep, x='Shooting Accuracy', y='Goals')
plt.title("Relacion entre Goles y precision");
```

```python
X = df_prep[['Shooting Accuracy']]
y = df_prep.Goals
```

```python
lr = LinearRegression()
pf = PolynomialFeatures()
svr = SVR(kernel='linear')

pipe_pr = Pipeline([('poly', PolynomialFeatures(2)),
                    ('lr', LinearRegression())])

ls_models = {'LinearRegression': lr, 'PolynomialRegression':pipe_pr, 'SVRLinear': svr}
ls_mae = []
ls_preds = []
for name, model in ls_models.items():
    model.fit(X, y)
    y_hat = model.predict(X)
    ls_preds.append(y_hat)

    mae = mean_absolute_error(y, y_hat)
    ls_mae.append(mae)

    print(name)
    print(f'MAE: {mae:.2f}')

    print("\n\n")
```

```python
df_res = X.copy()
df_res['Goals'] = y
for i, name in enumerate(ls_models.keys()):
    df_res[name] = ls_preds[i]

df_res
```

```python
lr_slope, lr_int = lr.coef_[0], lr.intercept_
print(f"La recta de regresion es goals = {lr_slope:.2f}*precision {lr_int:.2f}")
```

```python
coefs = pipe_pr['lr'].coef_
pr_slope_x = coefs[1]
pr_slope_x_2 = coefs[2]
pr_int = pipe_pr['lr'].intercept_

print(f"La parábola de regresion es goals = {pr_slope_x_2:.2f}*precision^2 {pr_slope_x:.2f}*precision + {pr_int:.2f}")
```

```python
svr_slope, svr_int = svr.coef_[0], svr.intercept_[0]
print(f"La recta de regresion SVR es goals = {svr_slope[0]:.2f}*precision + {svr_int:.2f}")
```

```python
x = np.linspace(20, 60)
sns.scatterplot(data=df_prep, x='Shooting Accuracy', y='Goals')
plt.plot(x, lr_slope * x + lr_int, 'r', label='LinearRegression')
plt.plot(x, pr_slope_x_2 * x**2 + pr_slope_x * x + pr_int, 'g', label='QuadraticRegression')
plt.plot(x, svr_slope * x + svr_int, 'k', label='LinearSVR')
plt.title("Relacion entre Goles y precision")
plt.legend();
```

## 5. Si tratamos la columna Goals como una columna categórica y usando logistic regression, SVM y Random forest, ¿cuántos goles marcaría un equipo que tiene...? (por cada algoritmo):

    - % de goles por disparo: 40%
    - Total shots (inc. Blocked): 102
    - Corners Taken: 5
    - Passes: 82
    - Dribbles: 23

```python
sorted(df.Goals.unique())
```

```python
df.Goals.hist();
```

```python
X = df_prep[[r'% Goals-to-shots', 'Total shots (inc. Blocked)', 'Corners Taken', 'Passes', 'Dribbles']]
y = df_prep['Goals']

lr = LogisticRegression()
tree = DecisionTreeClassifier(min_samples_leaf=2)
svc = SVC(kernel='linear', C=.005)

ls_models = {'LogisticRegression': lr, 'DecisionTreeClassifier':tree, 'SVCLinear': svc}
ls_acc = []
ls_preds = []
for name, model in ls_models.items():
    model.fit(X, y)
    y_hat = model.predict(X)
    ls_preds.append(y_hat)

    acc = accuracy_score(y, y_hat)
    ls_acc.append(acc)

    print(name)
    print(f'ACCURACY: {acc:.2%}')

    print("\n\n")
```

```python
sns.swarmplot(data=df_prep, x='Goals', hue=ls_preds[1], palette=['r', 'g', 'b', 'orange', 'k', 'brown'])
plt.title("Decision Tree");
```

```python
sns.swarmplot(data=df_prep, x='Goals', hue=ls_preds[0], palette=['r', 'g', 'b', 'orange', 'k', 'brown', 'pink'])
plt.title("Logistic Regression");
```

```python
sns.swarmplot(data=df_prep, x='Goals', hue=ls_preds[2], palette=['r', 'g', 'b', 'orange', 'k', 'brown', 'pink'])
plt.title("SVR Regression");
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

disp = ConfusionMatrixDisplay(confusion_matrix(y, ls_preds[0]))
disp.plot()
plt.title("LogisticRegression")
```
