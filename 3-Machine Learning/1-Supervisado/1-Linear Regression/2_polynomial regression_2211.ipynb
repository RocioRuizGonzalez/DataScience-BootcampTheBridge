# Polynomial regression
En muchas ocasiones la relación entre los datos no va a ser lineal, por lo que una simple regresión lineal no se va a ajustar bien a nuestra problemática.

Por ejemplo, en el siguiente ejmplo encajaria muy bien una regresión lineal.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data/salary_data.csv')
df.head()
```

```python
df.describe()
```

```python
sns.scatterplot(data=df, x='YearsExperience', y='Salary');
```

```python
sns.regplot(data=df, x='YearsExperience', y='Salary');
```

En la vida real los problemas son más complejos, ya que los salarios no llevan una subida lineal vs años de experiencia, dependerá también de la carrera profesional: profesor, ingeniero, doctor...

Otra posible representación del salario podría ser en función del puesto de trabajo.

```python
df = pd.read_csv('data/position_salaries.csv')
df
```

```python
X = df[['Level']]
y = df['Salary']
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
# Visualizing the Linear Regression results
def viz_linear():
    plt.scatter(X, y, color='red')
    plt.plot(X, lin_reg.predict(X), color='blue')
    plt.title('Truth or Bluff (Linear Regression)')
    plt.xlabel('Position level')
    plt.ylabel('Salary')
    plt.show()
    return
viz_linear()
```

Vemos que en este caso, la regresión lineal no se ajusta nada bien a los datos, por lo que habrá que probar otras alternativas.

`PolynomialFeatures` de sklearn genera nuevas features sintéticas, realizando todas las combinaciones posibles entre las potencias. Si degree = 3 y tenemos dos features (a, b), realizará a^2, b^2, a^3, b^3, ab, a^2b, ab^2

```python
from sklearn.preprocessing import PolynomialFeatures

pf = PolynomialFeatures(3, include_bias=False)
poly = pf.fit_transform(df[['Level',	'Salary']])
poly
```

```python
pf.powers_
```

```python
pd.DataFrame(poly, columns=['Level', 'Salary', 'Level^2', 'Salary*Level', 'Salary^2', 'Level^3', 'Level^2*Salary', 'Level*Salary^2', 'Salary^3'])
```

```python
poly_reg = PolynomialFeatures(6)
pol_reg = LinearRegression()

X_t = poly_reg.fit_transform(X)
pol_reg.fit(X_t, y) 
```

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 5))
ax = ax.reshape(-1)
for i, deg in enumerate(range(2, 6)):
    poly_reg = PolynomialFeatures(deg)
    pol_reg = LinearRegression()

    X_t = poly_reg.fit_transform(X)
    pol_reg.fit(X_t, y) 
    # Visualizing the Polymonial Regression results
    def viz_polymonial():
        ax[i].scatter(X, y, color='red')
        ax[i].plot(X, pol_reg.predict(poly_reg.transform(X)), color='blue')
        ax[i].set_title(f'Truth or Bluff (Linear Regression) (GRADO {deg})')
        ax[i].set_xlabel('Position level')
        ax[i].set_ylabel('Salary')
        # plt.show()
        return
    
    viz_polymonial()
    plt.tight_layout()
```

Al añadir mucha complejidad podemos sobreajustar el modelo fácilmente

```python
import numpy as np
np.random.seed(0)
df_train = df.sample(5)
df_test = df.sample(5)

X_train = df_train[['Level']]
X_test = df_test[['Level']]

y_train = df_train['Salary']
y_test = df_test['Salary']
```

```python
poly = PolynomialFeatures(2, include_bias=False)
lm = LinearRegression()

X_train_t = poly.fit_transform(X_train)
lm.fit(X_train_t, y_train)
```

```python
ls_mae = []
ls_mae_train = []
for deg in range(2, 11):
    print("#" * 50)
    print(f"Grado {deg}")
    poly = PolynomialFeatures(deg, include_bias=False)
    lm = LinearRegression()

    X_train_t = poly.fit_transform(X_train)
    lm.fit(X_train_t, y_train)

    y_hat = lm.predict(poly.transform(X_test))

    mae = np.mean(np.abs(y_test - y_hat))
    mae_train = np.mean(np.abs(y_train - lm.predict(X_train_t)))

    ls_mae.append(mae)
    ls_mae_train.append(mae_train)

    print("MAE TRAIN:", mae_train)
    print("MAE TEST:", mae)
    print("\n\n")
```

```python
x = range(2, 11)
plt.plot(x, ls_mae, label="MAE Test", color='r')
plt.plot(x, ls_mae_train, label="MAE Train", color='b')
plt.legend();
```
