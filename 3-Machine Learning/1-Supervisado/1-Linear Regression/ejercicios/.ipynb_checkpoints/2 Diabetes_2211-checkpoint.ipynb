## Buliding a regression model

The data have already been split to predictor (*data*) and response (*target*) variables. Given this information, we'll apply what we have previously learned about linear regression and apply the algorithm to the diabetes dataset.

#### In the cell below, import the `linear_model` class from `sklearn`. 

```python
from sklearn.linear_model import LinearRegression
```

#### Create a new instance of the linear regression model and assign the new instance to the variable `diabetes_model`.

```python
diabetes_model = LinearRegression()
```

#### Next, let's split the training and test data.

Define `diabetes_data_train`, `diabetes_target_train`, `diabetes_data_test`, and `diabetes_target_test`. Use the last 20 records for the test data and the rest for the training data.

```python
df.columns
```

```python
df_mod = df.copy()
df_mod = df_mod[['sex', 'bmi', 'bp', 's1', 's2','s3', 's4', 's5', 's6']]
X_train = df_mod.iloc[:-20]
X_test = df_mod.iloc[-20:]
y_train = target[:-20]
y_test = target[-20:]

print("TRAIN SIZES: ", X_train.shape, y_train.shape)
print("TEST SIZES: ", X_test.shape, y_test.shape)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
sns.histplot(y_train, alpha=0.5, bins=15, ax=ax)
sns.histplot(y_test, alpha=0.5,bins=15,  color='r', ax=ax);
```

```python
df_plot = df.copy()
df_plot['target'] = target
sns.pairplot(df_plot)
plt.tight_layout();
```

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df_plot.corr(), vmin=-1, vmax=1, annot=True);
```

Fit the training data and target to `diabetes_model`. Print the *intercept* and *coefficients* of the model.

```python
df.describe()
```

```python
df.hist()
plt.tight_layout()
```

```python
plt.hist(target)
```

```python
diabetes_model.fit(X_train, y_train)
```

#### Inspecting the results

From the outputs you should have seen:

- The intercept is a float number.
- The coefficients are an array containing 10 float numbers.

This is the linear regression model fitted to your training dataset.

#### Using your fitted linear regression model, predict the *y* of `diabetes_data_test`.

```python
y_hat = diabetes_model.predict(X_test)
y_hat
```

#### Print your `diabetes_target_test` and compare with the prediction. 

```python
y_test
```

#### Is `diabetes_target_test` exactly the same as the model prediction?

```python
res = y_test - y_hat
sns.scatterplot(res)
plt.title("Residuos");
```

```python
sns.histplot(res, bins=15)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error
import numpy as np

mae = mean_absolute_error(y_test, y_hat)
mape = mean_absolute_percentage_error(y_test, y_hat)
rmse = np.sqrt(mean_squared_error(y_test, y_hat))
r2 = r2_score(y_test, y_hat)

print("MAE:", mae)
print("MAPE:", mape)
print("RMSE:", rmse)
print("R2:", r2)
```

```python
np.mean(y_train)
```

#### Which are the most important features?

```python
df_res = pd.DataFrame(diabetes_model.coef_, index=X_train.columns, columns=["Coefs"])
df_res.sort_values(by='Coefs').plot(kind='barh')
```
