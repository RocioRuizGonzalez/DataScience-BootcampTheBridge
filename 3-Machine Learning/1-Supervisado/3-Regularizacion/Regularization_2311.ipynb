### 1. Cargamos librerías

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import datasets 
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
```

### 2. Vemos qué pinta tiene el dataset

```python
# Loading pre-defined Boston Dataset 
boston_dataset = datasets.load_boston() 
print(boston_dataset.DESCR) 
```

### 3. Rápido análisis exploratorio para ver sus dependencias lineales

```python
# Generate scatter plot of independent vs Dependent variable 
plt.style.use('ggplot') 
fig = plt.figure(figsize = (18, 18)) 
  
for index, feature_name in enumerate(boston_dataset.feature_names): 
    ax = fig.add_subplot(4, 4, index + 1) 
    ax.scatter(boston_dataset.data[:, index], boston_dataset.target) 
    ax.set_ylabel('House Price', size = 12) 
    ax.set_xlabel(feature_name, size = 12) 
  
plt.show() 
```

### 4. Preprocesado de los datos

```python
# Load the dataset into Pandas Dataframe 
boston_pd = pd.DataFrame(boston_dataset.data) 
boston_pd.columns = boston_dataset.feature_names 
boston_pd_target = np.asarray(boston_dataset.target) 
boston_pd['House Price'] = pd.Series(boston_pd_target) 

boston_pd
```

```python
 
# input  
X = boston_pd.iloc[:, :-1] 
  
#output 
Y = boston_pd.iloc[:, -1] 
```

```python
print(boston_pd.head())

x_train, x_test, y_train, y_test = train_test_split( 
    X, Y,  
    test_size = 0.25) 
  
print("Train data shape of X = % s and Y = % s : "%( 
    x_train.shape, y_train.shape)) 
  
print("Test data shape of X = % s and Y = % s : "%( 
    x_test.shape, y_test.shape)) 
```

### 5. Regresión lineal múltiple

```python
# Apply multiple Linear Regression Model 
### CODE
from sklearn.preprocessing import StandardScaler

lreg = LinearRegression()
sc = StandardScaler()

x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)

lreg.fit(x_train_sc, y_train)
lreg_y_pred = lreg.predict(x_test_sc)


# calculating Mean Squared Error (mse) 
mean_squared_error = np.mean((lreg_y_pred - y_test)**2) 
print("Mean squared Error on test set : ", mean_squared_error) 
  
# Putting together the coefficient and their corrsponding variable names  
lreg_coefficient = pd.DataFrame() 
lreg_coefficient["Columns"] = x_train.columns 
lreg_coefficient['Coefficient Estimate'] = pd.Series(lreg.coef_) 
print(lreg_coefficient) 
```

```python
print("Coeficientes:", lreg.coef_)
print("Interceptor:", lreg.intercept_)
```

### 6. Ridge regression

```python
# import ridge regression from sklearn library 
#### CODE
from sklearn.linear_model import Ridge
ridgeR = Ridge(alpha=1.)

sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)

ridgeR.fit(x_train_sc, y_train)
y_pred = ridgeR.predict(x_test_sc)

# calculate mean square error 
mean_squared_error_ridge = np.mean((y_pred - y_test)**2) 
print(mean_squared_error_ridge) 
  
# get ridge coefficient and print them 
ridge_coefficient = pd.DataFrame() 
ridge_coefficient["Columns"]= x_train.columns 
ridge_coefficient['Coefficient Estimate'] = pd.Series(ridgeR.coef_) 
print(ridge_coefficient) 
```

```python
print("Coeficientes:", ridgeR.coef_)
print("Interceptor:", ridgeR.intercept_)
```

Veamos en la gráfica cómo quedaría para alpha = 1

```python
def plot_params(model, columns):
    df_res = pd.DataFrame(model.coef_, index=columns, columns=['Coefs'])
    df_res.sort_values(by='Coefs').plot(kind='barh')

plot_params(ridgeR, x_train.columns)
```

Si hacemos ahora alpha = 10

```python
# Train the model  
ridgeR = Ridge(alpha=100.)

sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)

ridgeR.fit(x_train_sc, y_train)
y_pred = ridgeR.predict(x_test_sc)
  
# calculate mean square error 
mean_squared_error_ridge = np.mean((y_pred - y_test)**2) 
print(mean_squared_error_ridge) 

mean_absolute_error_ridge = np.mean(np.abs(y_pred - y_test)) 
print('MAE:', mean_absolute_error_ridge) 
  
# get ridge coefficient and print them 
ridge_coefficient = pd.DataFrame() 
ridge_coefficient["Columns"]= x_train.columns 
ridge_coefficient['Coefficient Estimate'] = pd.Series(ridgeR.coef_) 
print(ridge_coefficient) 

# plotting the coefficient score 
plot_params(ridgeR, x_train.columns)
```

### 7. Lasso regression

```python
# import Lasso regression from sklearn library 
#### CODE

from sklearn.linear_model import Lasso
# Train the model  
lasso = Lasso(alpha=.1)

sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)

lasso.fit(x_train_sc, y_train)
y_pred1 = lasso.predict(x_test_sc)

# Calculate Mean Squared Error 
mean_squared_error = np.mean((y_pred1 - y_test)**2) 
print("Mean squared error on test set", mean_squared_error) 
mean_absolute_error= np.mean(np.abs(y_pred1 - y_test)) 
print('MAE:', mean_absolute_error) 
  

lasso_coeff = pd.DataFrame() 
lasso_coeff["Columns"] = x_train.columns 
lasso_coeff['Coefficient Estimate'] = pd.Series(lasso.coef_) 
  
print(lasso_coeff) 

# plotting the coefficient score 
# plotting the coefficient score 
plot_params(lasso, x_train.columns)
```

### 8. Elastic Net

```python
# import Lasso regression from sklearn library 
#### CODE

from sklearn.linear_model import ElasticNet
# Train the model  
elastic = ElasticNet(alpha=1., l1_ratio=0.1)

sc = StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)

elastic.fit(x_train_sc, y_train)
y_pred1 = elastic.predict(x_test_sc)

# Calculate Mean Squared Error 
mean_squared_error = np.mean((y_pred1 - y_test)**2) 
print("Mean squared error on test set", mean_squared_error) 
mean_absolute_error= np.mean(np.abs(y_pred1 - y_test)) 
print('MAE:', mean_absolute_error) 
  

lasso_coeff = pd.DataFrame() 
lasso_coeff["Columns"] = x_train.columns 
lasso_coeff['Coefficient Estimate'] = pd.Series(lasso.coef_) 
  
print(lasso_coeff) 

# plotting the coefficient score 
# plotting the coefficient score 
plot_params(elastic, x_train.columns)
```

Conclusion :
From the above analysis we can reach the following conclusion about different regularization methods:

Regularization is used to reduce the dependence on any particular independent variable by adding the penalty term to the Loss function. This term prevents the coefficients of the independent variables to take extreme values.
Ridge Regression adds L2 regularization penalty term to loss function. This term reduces the coefficients but does not make them 0 and thus doesn’t eliminate any independent variable completely. It can be used to measure the impact of the different independent variables.
Lasso Regression adds L1 regularization penalty term to loss function. This term reduces the coefficients as well as makes them 0 thus effectively eliminate the corresponding independent variable completely. It can be used for feature selection etc.
Elastic Net is a combination of both of the above regularization. It contains both the L1 and L2 as its penalty term. It performs better than Ridge and Lasso Regression for most of the test cases.

## Método 1: Train, Valid, Test

```python
X_trainval, X_test, y_trainval, y_test = train_test_split(X,
                                                    Y,
                                                    train_size=.8,
                                                    random_state=0)

print(f"TRAIN + VALID: {X_trainval.shape} - {y_trainval.shape}")
print(f"TEST: {X_test.shape} - {y_test.shape}")
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_trainval,
                                                    y_trainval,
                                                    train_size=.8,
                                                    random_state=0)

print(f"TRAIN: {X_train.shape} - {y_train.shape}")
print(f"VAL: {X_val.shape} - {y_val.shape}")
```

```python
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_val_sc = sc.transform(X_val)

X_trainval_sc = sc.fit_transform(X_trainval)
X_test_sc = sc.transform(X_test)
```

```python
grid_alpha = np.linspace(0, 1, 10)
grid_l1 = np.linspace(0, 1, 10)

d_mae = {'alpha':[], 'l1_ratio':[], 'mae':[]}
for alp in grid_alpha:
    for r in grid_l1:
        elastic = ElasticNet(alpha=alp, l1_ratio=r)

        elastic.fit(X_train_sc, y_train)
        y_pred1 = elastic.predict(X_val_sc)

        mae = np.mean(np.abs(y_val - y_pred1))

        d_mae['alpha'].append(alp)
        d_mae['l1_ratio'].append(r)
        d_mae['mae'].append(mae)
```

```python
df_mae = pd.DataFrame(d_mae)
```

```python
fig, ax = plt.subplots()
sns.scatterplot(x=d_mae['alpha'], y=d_mae['l1_ratio'], c=(np.array(d_mae['mae']) * 50), ax=ax)
plt.xlabel('alpha')
plt.ylabel('l1_ratio')
```

```python
grid_alpha = np.linspace(0, 150, 10)

d_mae = {'alpha':[], 'mae':[], 'mae_train':[]}
for alp in grid_alpha:
    model = Ridge(alpha=alp)

    model.fit(X_train_sc, y_train)
    y_pred1 = model.predict(X_val_sc)
    y_pred2 = model.predict(X_train_sc)

    mae = np.mean(np.abs(y_val - y_pred1))
    mae_train = np.mean(np.abs(y_train - y_pred2))

    d_mae['alpha'].append(alp)
    d_mae['mae'].append(mae)
    d_mae['mae_train'].append(mae_train)

df_mae = pd.DataFrame(d_mae)
```

```python
plt.plot(d_mae['alpha'], d_mae['mae'], '*r', label='Validation')
plt.plot(d_mae['alpha'], d_mae['mae_train'], 'sb', label='Train')
plt.legend()
```
