# Preparacion para entrenamiento

```python
from sklearn.model_selection import train_test_split

X = data.copy()
X = X.drop(columns='sales')
y = data['sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=.8, random_state=0)

print("TRAIN: {} - {}".format(X_train.shape, y_train.shape))
print("TEST: {} - {}".format(X_test.shape, y_test.shape))
```

```python
fig, ax = plt.subplots(1, 2)
sns.histplot(y_train, ax=ax[0])
sns.histplot(y_test, ax=ax[1])
ax[0].set_title('Train Sales')
ax[1].set_title('Test Sales')
fig.tight_layout()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, mean_absolute_percentage_error
def metrics_report(y_true, y_pred, verbosity=True):
    rmse = mean_squared_error(y_true, y_pred) ** 0.5
    mae = mean_absolute_error(y_true, y_pred)
    mape = mean_absolute_percentage_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    if verbosity:
        print(f"RMSE: {rmse}")
        print(f"MAE: {mae}")
        print(f"MAPE: {mape}")
        print(f"R2: {r2}")

    return rmse, mae, mape, r2
```

```python
def train_prep(X_train, transf):
    X_train_prep = X_train.copy()
    
    ## Feature Engineering
    # X_train_prep['TV_Radio'] = X_train_prep['TV'] * X_train_prep['radio']

    ## Transformaciones
    transf_out = []
    for t in transf:
        X_train_prep = t.fit_transform(X_train_prep)
        transf_out.append(t)

    return X_train_prep, transf_out

def test_prep(X_test, transf):

    X_test_prep = X_test.copy()

    ## Feature Engineering
    # X_test_prep['TV_Radio'] = X_test_prep['TV'] * X_test_prep['radio']

    ## Transformaciones
    for t in transf:
        X_test_prep = t.transform(X_test_prep)

    return X_test_prep

def train_test_model(model, X_train, X_test, y_train, y_test, name=None):
    if name is not None:
        print(f"\n{name.upper()}")
        
    model.fit(X_train, y_train)
    y_hat = model.predict(X_test)

    metrics_report(y_test, y_hat)

    return model,  y_hat
```

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet, LinearRegression
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
transf = [sc]

X_train_prep, transf_out = train_prep(X_train, transf)
X_test_prep = test_prep(X_test, transf_out)

# print("\nLINEAR REGRESSION")
lm = LinearRegression()
lm, y_hat_lm = train_test_model(lm, X_train_prep, X_test_prep, y_train, y_test, name='Linear Regression')

print("\nRIDGE REGRESSION")
ridge = Ridge(alpha=.3)
ridge, y_hat_ridge = train_test_model(ridge, X_train_prep, X_test_prep, y_train, y_test)

print("\nLASSO REGRESSION")
lasso = Lasso(alpha=.3)
lasso, y_hat_lasso = train_test_model(lasso, X_train_prep, X_test_prep, y_train, y_test)

print("\nElasticNet REGRESSION")
en = ElasticNet(alpha=.3, l1_ratio=.5)
en, y_hat_en = train_test_model(en, X_train_prep, X_test_prep, y_train, y_test)
```

```python
d_models = {'Linear Regression': LinearRegression(),
                'Ridge': Ridge(alpha=.3),
                'Lasso': Lasso(alpha=.3)}

d_out = {}
for name, mod in d_models.items():
    d_out[name] = {'model':None}
    model, y_hat = train_test_model(mod, X_train_prep, X_test_prep, y_train, y_test, name=name)
    d_out[name]['model']= model
    # d_out[name].append(y_hat)
    rmse, mae, mape, r2 = metrics_report(y_test, y_hat, verbosity=False)
    d_out[name]['RMSE'] = rmse
    d_out[name]['MAE'] = mae
    d_out[name]['MAPE'] = mape
    d_out[name]['R2'] = r2
    
```

```python
df_res = pd.DataFrame(d_out).T
df_res.sort_values(by='MAE')
```
