```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/train.csv')
```

```python
df
```

```python
X = df[['Pulgadas', 'Tipo']]
y = df['Precio']
```

```python
X = pd.get_dummies(X, columns=['Tipo'], drop_first=True)
X
```

```python
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
lm.fit(X, y)
```

```python
from sklearn.model_selection import cross_val_score

error_modelo_1 = np.mean(cross_val_score(lm, X, y, scoring='neg_root_mean_squared_error'))
error_modelo_1
```

```python
df_test = pd.read_csv('data/test.csv')
```

```python
df_test
```

```python
X_test = df_test[['Pulgadas', 'Tipo']]
X_test = pd.get_dummies(X_test, columns=['Tipo'], drop_first=True)
```

```python
X_test
```

```python
y_test = lm.predict(X_test)
```

```python
df_sub = pd.DataFrame({'Id': df_test.Id, 'Precio':y_test})
df_sub
```

## Probamos otras variables que puedan mejorar el modelo

### Analizando precios medios segÃºn fabricantes

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
visplot = df.groupby(by='Fabricante')[list(['Fabricante', 'Precio'])].mean()
visplot.reset_index(inplace=True)
chart = sns.barplot(data=visplot, x='Fabricante', y='Precio', width=0.8)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

El precio medio varÃ­a bastante en funciÃ³n del fabricante, por lo que podemos pensar que introducir el precio en nuestro modelo puede ayudar a hacer mejores predicciones.

### Creando nuevo modelo que tenga en cuenta el fabricante

```python
X = df[['Pulgadas', 'Tipo', 'Fabricante']]
X = pd.get_dummies(X, columns=['Tipo'], drop_first=True)
X = pd.get_dummies(X, columns=['Fabricante'], drop_first=True)
```

```python
lm = LinearRegression()
lm.fit(X, y)
```

```python
error_modelo_2 = np.mean(cross_val_score(lm, X, y, scoring='neg_root_mean_squared_error'))
print(f"Error del primer modelo: {error_modelo_1}")
print(f"Error del segundo modelo: {error_modelo_2}")
```

*Â¡El nuevo modelo mejora los resultados del primero!*

```python
X_test = df_test[['Pulgadas', 'Tipo', 'Fabricante']]
X_test['Fabricante'].unique()
X_test = pd.get_dummies(X_test, columns=['Tipo'], drop_first=True)
X_test = pd.get_dummies(X_test, columns=['Fabricante'], drop_first=True)
# AÃ±adimos columnas con valor cero para los valores que no se encontraban en el conjunto de test
X_test['Fabricante_Google'] = 0
X_test['Fabricante_LG'] = 0
X_test['Fabricante_Vero'] = 0
# Reordenamos las columnas de X_test para que tengan el mismo orden que en X
X_test = X_test[X.columns.values]
y_test = lm.predict(X_test)
```

```python
df_sub = pd.DataFrame({'Id': df_test.Id, 'Precio':y_test})
df_sub.to_csv('submissions/submission_2.csv', index=False)
```

### Analizamos los precios medios segÃºn memoria RAM

```python
plt.figure(figsize=(10,5))
visplot = df.groupby(by='RAM')[list(['RAM', 'Precio'])].mean()
visplot.reset_index(inplace=True)
chart = sns.barplot(data=visplot, x='RAM', y='Precio', width=0.8)
chart.set_xticklabels(chart.get_xticklabels(), rotation=45);
```

El precio medio varÃ­a bastante en funciÃ³n del tamaÃ±o de la memoria RAM, por lo que podemos pensar que introducir el precio en nuestro modelo puede ayudar a hacer mejores predicciones.

```python
lm = LinearRegression()
lm.fit(X, y)
```

```python
error_modelo_5 = np.mean(cross_val_score(lm, X, y, scoring='neg_root_mean_squared_error'))
print(f"Error del primer modelo: {error_modelo_1}")
print(f"Error del segundo modelo: {error_modelo_2}")
print(f"Error del tercer modelo: {error_modelo_3}")
print(f"Error del cuarto modelo: {error_modelo_4}")
print(f"Error del quinto modelo: {error_modelo_5}")
```

### Creando un nuevo modelo que tenga en cuenta si la pantalla es tÃ¡ctil

```python
X = df[['Pulgadas', 'Tipo', 'Fabricante', 'RAM', 'Resolution', 'ScreenResolution']]
X = pd.get_dummies(X, columns=['Tipo'], drop_first=True)
X = pd.get_dummies(X, columns=['Fabricante'], drop_first=True)
memoria = {'8GB': 8, '4GB': 4, '6GB': 6, '12GB': 12, '16GB': 16, '32GB': 32, '2GB': 2, '24GB': 24, '64GB': 64}
X['RAM'] = X['RAM'].map(memoria)
X['Touchscreen'] = X.ScreenResolution.str.contains('Touchscreen')
X['Touchscreen'] = X['Touchscreen'].astype(int)
X.drop('ScreenResolution', inplace=True, axis=1)
```

```python
lm = LinearRegression()
lm.fit(X, y)
```

```python
error_modelo_6 = np.mean(cross_val_score(lm, X, y, scoring='neg_root_mean_squared_error'))
print(f"Error del primer modelo: {error_modelo_1}")
print(f"Error del segundo modelo: {error_modelo_2}")
print(f"Error del tercer modelo: {error_modelo_3}")
print(f"Error del cuarto modelo: {error_modelo_4}")
print(f"Error del quinto modelo: {error_modelo_5}")
print(f"Error del sexto modelo: {error_modelo_6}")
```

*Â¡El nuevo modelo no mejora los resultados del quinto!*
