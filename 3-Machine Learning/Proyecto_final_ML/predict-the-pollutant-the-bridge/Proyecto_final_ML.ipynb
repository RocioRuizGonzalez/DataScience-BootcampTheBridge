## Definimos un primer modelo con LinearSVC(C=1, loss='hinge', max_iter=10000)

```python
X = df[["eprtrSectorName","avg_wind_speed"]]
y = df["pollutant"]
```

```python
X = pd.get_dummies(X, columns=['eprtrSectorName'], drop_first=True)
X
```

```python
from sklearn.svm import LinearSVC

svc1 = LinearSVC(C=1, loss='hinge', max_iter=10000)

svc1.fit(X, y)
```

```python
df_test = pd.read_csv('data/test.csv',sep=",")
df_test.head()
```

```python
X_test = df_test[["eprtrSectorName","avg_wind_speed"]]
X_test = pd.get_dummies(X_test, columns=['eprtrSectorName'], drop_first=True)
```

```python
X_test.head()
```

```python
y_test = svc1.predict(X_test)
y_test
```

```python
df_test["y_test"]=y_test
df_test.head()
```

```python
from sklearn.metrics import accuracy_score

y_hat = svc1.predict(X)
acc = accuracy_score(y_true=y, y_pred=y_hat)

print("ACCURACY:", acc)
```

```python
df_sub = pd.DataFrame({'Id': df_test.index, 'pollutant':y_test})
df_sub
```
