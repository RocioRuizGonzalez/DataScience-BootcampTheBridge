## Predicciones
1. Divide en train y test
2. Prepara un pipeline compuesto por:
    - StandardScaler,
    - PCA de 5 componentes
    - KNN con 3 vecinos
3. Entrena
4. Predice en test
5. Calcula el accuracy score en train y test
6. Representa la matriz de confusión
7. ¿Qué combinación de vecinos y componentes mejoraría el accuracy en test?

```python
X = df[["ingresos","gastos_comunes","pago_coche","gastos_otros","ahorros","vivienda","estado_civil","hijos","trabajo"]]
y = df["comprar"]

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y,
                                                    test_size=0.2,
                                                    random_state=0)
```

```python
scaler = StandardScaler()
pca = PCA(n_components=5)
knn=KNeighborsClassifier(n_neighbors=3)


pipe = Pipeline(steps=[("scaler", scaler), ("pca", pca), ("knn", knn)])

pipe.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
y_hat = pipe.predict(X_test)
accu = accuracy_score(y_true=y_test, y_pred=y_hat)
accu
```

```python
from yellowbrick.classifier import ConfusionMatrix

cm = ConfusionMatrix(pipe)
cm.fit(X_train, y_train)
cm.score(X_test, y_test)
cm.show();
```

Mejoraría el accuracy si subimos los componentes de PCA y el número de vecinos. El óptimo sería 5 vecinos y 6 componentes de PCA
