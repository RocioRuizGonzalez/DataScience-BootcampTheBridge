9. Bas치ndote en el gr치fico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los m칠todos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python
fig, ax=plt.subplots()
bars = plt.bar(x=["Ana","Luis","Pedro","Sofia","Carmen"],height=[160,180,175,190,170],color="g")
for bar in bars:
    height = bar.get_height()
    x = bar.get_x()
    width = bar.get_width()
    offset = width / 3
    plt.text(x + offset, height + 1, height, va='bottom')
plt.show()


```

10. Cambia el tama침o de la figura a (3,5)

```python
fig, ax=plt.subplots()
bars = plt.bar(x=["Ana","Luis","Pedro","Sofia","Carmen"],height=[160,180,175,190,170],color="g")
for bar in bars:
    height = bar.get_height()
    x = bar.get_x()
    width = bar.get_width()
    offset = width / 3
    plt.text(x + offset, height + 1, height, va='bottom')
plt.rcParams["figure.figsize"] = [3,5]
plt.show()
```
