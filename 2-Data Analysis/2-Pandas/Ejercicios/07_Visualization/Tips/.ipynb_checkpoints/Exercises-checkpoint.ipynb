### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

### Step 3. Assign it to a variable called tips

```python
url="https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv"
tips=pd.read_csv(url,sep=",")
tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop(["Unnamed: 0"], axis=1)
```

### Step 5. Plot the total_bill column histogram

```python
total_bill_histogram=tips["total_bill"].hist(bins=30, grid=True, rwidth=0.9,color='#607c8e')
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
tips.plot(x="total_bill",y="tip",kind="scatter");
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

### Step 8. Present the relationship between days and total_bill value

```python
tips.plot(x="day",y="total_bill",kind="scatter");
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.scatter(x=tips[tips["sex"]=="Female"]["tip"],y=tips[tips["sex"]=="Female"]["day"],c="r", label="Female")
ax1.scatter(x=tips[tips["sex"]=="Male"]["tip"],y=tips[tips["sex"]=="Male"]["day"],c="g", label="Male")
plt.legend(loc='best')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
tips.plot(x="day",y="total_bill", kind="box")
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
plt.subplots(1,2)
plt.subplot(1, 2, 1)
tips[tips["time"]=="Lunch"]["tip"].hist(bins=30, grid=True, rwidth=0.9,color='#607c8e')
plt.legend("Lunch")
plt.subplot(1, 2, 2)
tips[tips["time"]=="Dinner"]["tip"].hist(bins=30, grid=True, rwidth=0.9,color='#607c8e')
plt.legend("Dinner")
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
plt.subplots(1,2)
plt.subplot(1, 2, 1)
tips[tips["smoker"]=="Lunch"]["tip"].scatter(bins=30, grid=True, rwidth=0.9,color='#607c8e')
plt.legend("Lunch")
plt.subplot(1, 2, 2)
tips[tips["time"]=="Dinner"]["tip"].hist(bins=30, grid=True, rwidth=0.9,color='#607c8e')
plt.legend("Dinner")
```

```python
plt.subplots(1,2)


plt.subplot(1, 2, 1)
female = tips[tips["sex"]=="Female"]
plt.scatter(x=female[female["smoker"]=="Yes"]["total_bill"],y=female[female["smoker"]=="Yes"]["tip"],c="r", label="Female smokers")
plt.scatter(x=female[female["smoker"]=="No"]["total_bill"],y=female[female["smoker"]=="No"]["tip"],c="b", label="Female non-smokers")
plt.legend(loc='best')
plt.tight_layout()

plt.subplot(1, 2, 2)
male = tips[tips["sex"]=="Male"]
plt.scatter(x=male[male["smoker"]=="Yes"]["total_bill"],y=male[male["smoker"]=="Yes"]["tip"],c="r", label="Male smokers")
plt.scatter(x=male[male["smoker"]=="No"]["total_bill"],y=male[male["smoker"]=="No"]["tip"],c="b", label="Male non-smokers")
plt.legend(loc='best')
plt.tight_layout()

plt.show()
```
